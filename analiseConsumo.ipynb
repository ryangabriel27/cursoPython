```python

import numpy as np
import matplotlib.pyplot as plt

# Valores fixos
valorGasolina = 6.20
valorEtanol = 4.12

# Entrada do usuário (pode ser convertida em float para cálculos)
distanciaGasolina = float(input("Distância que o carro percorre com 1L de gasolina: "))
distanciaEtanol = float(input("Distância que o carro percorre com 1L de etanol: "))

# Custo por km
custo_km_gasolina = valorGasolina / distanciaGasolina
custo_km_etanol = valorEtanol / distanciaEtanol
```

```python
# Gráfico 1: Comparação do custo por km
plt.figure(figsize=(10, 5))
plt.bar(['Gasolina', 'Etanol'], [custo_km_gasolina, custo_km_etanol], color=['blue', 'green'])
plt.title('Custo por km rodado')
plt.ylabel('R$/km')
plt.grid(axis='y')
plt.show()
```

```python
# Gráfico 2: Quando a gasolina compensa mais que o etanol
eficiencia_relativa = distanciaEtanol / distanciaGasolina
preco_relativo = valorEtanol / valorGasolina

plt.figure(figsize=(10, 5))
plt.axhline(0.70, color='red', linestyle='--', label='Limite de vantagem da gasolina (70%)')
plt.plot([0, 1], [preco_relativo]*2, label='Relação de preço Etanol/Gasolina', color='green')
plt.plot([0, 1], [eficiencia_relativa]*2, label='Relação de rendimento Etanol/Gasolina', color='blue')
plt.legend()
plt.title('Análise de Vantagem Econômica')
plt.ylabel('Proporção Etanol/Gasolina')
plt.xticks([])
plt.show()
```

```python
# Eixo X: km rodados (de 1 a 1000 km)
kms = np.linspace(1, 1000, 1000)

# Custo total para cada combustível
custo_gasolina = (kms / distanciaGasolina) * valorGasolina
custo_etanol = (kms / distanciaEtanol) * valorEtanol

# Encontrar o ponto de interseção
diferenca = custo_etanol - custo_gasolina
indice_cruzamento = np.where(diferenca > 0)[0]
km_cruzamento = kms[indice_cruzamento[0]] if len(indice_cruzamento) > 0 else None

# Plot
plt.figure(figsize=(10, 6))
plt.plot(kms, custo_gasolina, label='Gasolina', color='blue')
plt.plot(kms, custo_etanol, label='Etanol', color='green')
if km_cruzamento:
    plt.axvline(km_cruzamento, color='red', linestyle='--', label=f'Ponto de compensação ≈ {km_cruzamento:.0f} km')
plt.title('Custo acumulado por distância')
plt.xlabel('Km rodado')
plt.ylabel('Custo total (R$)')
plt.legend()
plt.grid(True)
plt.show()

if km_cruzamento:
    print(f"\n➡ A gasolina se torna mais econômica a partir de aproximadamente {km_cruzamento:.0f} km.")
else:
    print("\nℹ O etanol continua sendo mais econômico em toda a faixa de km considerada.")
```

```python
# --- Bloco adicional: cálculo de preços limiares ---

# Preço máximo da gasolina para que etanol seja mais vantajoso:
# custo_km_etanol < custo_km_gasolina
# => valorEtanol / distanciaEtanol < valorGasolina_limite / distanciaGasolina
valorGasolina_limite = (valorEtanol / distanciaEtanol) * distanciaGasolina
print(f"Para o etanol ser vantajoso, o preço da gasolina deve ser maior que R$ {valorGasolina_limite:.2f} por litro.")

# Preço máximo do etanol para que gasolina seja mais vantajosa:
# custo_km_gasolina < custo_km_etanol
# => valorGasolina / distanciaGasolina < valorEtanol_limite / distanciaEtanol
valorEtanol_limite = (valorGasolina / distanciaGasolina) * distanciaEtanol
print(f"Para a gasolina ser vantajosa, o preço do etanol deve ser maior que R$ {valorEtanol_limite:.2f} por litro.")
```
